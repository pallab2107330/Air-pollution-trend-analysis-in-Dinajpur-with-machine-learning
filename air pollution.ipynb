```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
data = {
    "Location": ["Hospital", "Residential", "Commercial", "Intersection", "Industrial"],
    "PM2.5": [77, 78.5, 120, 113.25, 95],
    "PM10": [110, 105, 130, 140, 115]
}
df = pd.DataFrame(data)

# Classify PM2.5 levels
def classify_pm25(pm):
    if pm <= 15:
        return "Good"
    elif pm <= 40:
        return "Moderate"
    elif pm <= 65:
        return "Unhealthy for Sensitive Groups"
    elif pm <= 150:
        return "Unhealthy"
    else:
        return "Very Unhealthy"

df["PM2.5_Quality"] = df["PM2.5"].apply(classify_pm25)

# --------- Analysis ---------
print("📊 Summary Statistics:\n", df.describe())
print("\n🛑 PM2.5 Air Quality Categories:\n", df["PM2.5_Quality"].value_counts())
print("\n🔥 Most Polluted:", df.loc[df["PM2.5"].idxmax(), "Location"])
print("🌿 Cleanest:", df.loc[df["PM2.5"].idxmin(), "Location"])
print("\n📈 Correlation (PM2.5 vs PM10):", df["PM2.5"].corr(df["PM10"]))

# --------- Graphs ---------

# Bar plot for PM2.5 and PM10
plt.figure(figsize=(10, 6))
df.plot(x="Location", y=["PM2.5", "PM10"], kind="bar", figsize=(10, 6))
plt.title("PM2.5 and PM10 Levels by Location")
plt.ylabel("µg/m³")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(x="PM2.5", y="PM10", data=df)
plt.title("PM2.5 vs PM10")
plt.xlabel("PM2.5 (µg/m³)")
plt.ylabel("PM10 (µg/m³)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Pie chart of PM2.5 Quality Categories
plt.figure(figsize=(6, 6))
df["PM2.5_Quality"].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Air Quality Classification Based on PM2.5")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
